## Vetores

Uma das maneiras mais comuns que utilizamos para agrupar dados é na forma de listas. Fazemos
listas de compras nos supermercados, listas de afazeres, listas de objetos para colocar na mala. Da
mesma maneira, uma das formas mais simples de armazenar dados de um tipo é através de listas,
aqui denominadas Vetores. Conforme definido anteriormente, vetor é um tipo de dado homogêneo,
portanto, todos os dados armazenados devem ser do mesmo tipo
